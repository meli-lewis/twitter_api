```python
import json
import os
from TwitterAPI import TwitterAPI
```

```python
api = TwitterAPI(consumer_key=os.getenv(key='TW_KEY'),
                 consumer_secret=os.getenv(key='TW_SECRET'),
                 access_token_key=os.getenv(key='TW_TOKEN'),
                 access_token_secret=os.getenv(key='TW_TOKEN_SECRET'))
```

```python
r = api.request('search/tweets', {'q':'cats',
                                  'lang':'en',
                                  'filter':'media'})
```
